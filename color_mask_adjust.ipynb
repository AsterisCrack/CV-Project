## Notebook to adjust color masks

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import sys, os, glob
```

```python
#Load images
root = os.getcwd()
path = os.path.join(root, "tracker_examples")
filenames = [i for i in glob.glob(path + '/green.jpg')]
file = filenames[0]
```

```python
# Read image
img = cv2.imread(file)
result = img.copy()
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

#Show image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_HSV2RGB))
```

```python
#Bounds in HSV. Red in this case
# lower boundary RED color range values; Hue (0 - 10)
lower1 = np.array([0, 100, 20])
upper1 = np.array([10, 255, 255])
 
# upper boundary RED color range values; Hue (160 - 180)
lower2 = np.array([160,100,20])
upper2 = np.array([179,255,255])
 
lower_mask = cv2.inRange(img, lower1, upper1)
upper_mask = cv2.inRange(img, lower2, upper2)
 
full_mask = lower_mask + upper_mask;
 
result = cv2.bitwise_and(result, result, mask=full_mask)
 
#Sublop with lower and upper mask
fig, ax = plt.subplots(1, 3, figsize=(20, 10))
ax[0].imshow(cv2.cvtColor(lower_mask, cv2.COLOR_BGR2RGB))
ax[1].imshow(cv2.cvtColor(upper_mask, cv2.COLOR_BGR2RGB))
ax[2].imshow(cv2.cvtColor(full_mask, cv2.COLOR_BGR2RGB))
plt.show()

#Show image
plt.imshow(cv2.cvtColor(full_mask, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()


```

```python
#For a single mask
lower = np.array([45, 100, 20])
upper = np.array([76, 255, 255])
mask = cv2.inRange(img, lower, upper)
result = cv2.bitwise_and(result, result, mask=mask)
plt.imshow(cv2.cvtColor(mask, cv2.COLOR_BGR2RGB))
plt.show()
```
