## Notebook to adjust color masks

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import sys, os, glob
```

```python
#Load images
root = os.getcwd()
path = os.path.join(root, "tracker_examples")
filenames = [i for i in glob.glob(path + '/red_ball.jpg')]
file = filenames[0]
```

```python
# Read image
img = cv2.imread(file)
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

#Show image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_HSV2RGB))
```

```python
#Bounds in HSV. Red in this case
lower_bound = (0, 100, 100)
upper_bound = (10, 255, 255)

#Show both colors as new images
def create_blank(width, height, rgb_color=(0, 0, 0)):
    """Create new image(numpy array) filled with certain color in RGB"""
    # Create black blank image
    image = np.zeros((height, width, 3), np.uint8)

    # Since OpenCV uses BGR, convert the color first
    color = tuple(reversed(rgb_color))
    # Fill image with color
    image[:] = color

    return image

# Create new blank 300x300 red image
lower = create_blank(300, 300, rgb_color=lower_bound)
upper = create_blank(300, 300, rgb_color=upper_bound)

#Show both colors as new images
plt.imshow(cv2.cvtColor(lower, cv2.COLOR_HSV2RGB))
plt.show()
plt.imshow(cv2.cvtColor(upper, cv2.COLOR_HSV2RGB))
```
